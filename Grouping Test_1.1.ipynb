```python
import pandas as pd
import numpy as np
```

```python
# Create data
columns = ['item', 'groupID']
df = pd.DataFrame(columns=columns)
df
```

```python
legumesGroup = ['Alfalfa sprouts', 'Azuki beans', 'Bean sprouts', 'Black beans', 'Black-eyed peas', 'Borlotti bean',
'Broad beans',
'Chickpeas', 'Garbanzos',
'Green beans',
'Kidney beans',
'Lentils',
'Lima beans',
'Mung beans',
'Navy beans',
'Pinto beans',
'Runner beans',
'Split peas',
'Soy beans',
'Peas']
```

```python
herbGroup = [
    'Anise',
'Basil',
'Caraway',
'Cilantro',
'Chamomile',
'Dill',
'Fennel',
'Lavender',
'Lemon Grass',
'Marjoram',
'Oregano',
'Parsley',
'Rosemary',
'Sage',
'Thyme'
]
```

```python
radishGroup = [
    'Rutabaga',
'Turnip',
'Wasabi',
'Horseradish',
'White radish'
]
```

```python
import random
random.choice(legumesGroup + herbGroup + radishGroup)
```

```python
groups = list()
groups.append(legumesGroup)
groups.append(herbGroup)
groups.append(radishGroup)
groups
```

```python
itemList = list()
for i in range(1000):
    itemList.append(random.choice(legumesGroup + herbGroup + radishGroup))
```

```python
itemList
```

```python
df['item'] = itemList
```

```python
df
```

```python
groupings = list()

for row in df.iterrows():
    for group in groups:
        print("data:", row[1][0])
        if row[1][0] in group:
#             row[1][10] = groups.index(group)
            print(groups.index(group))
            groupings.append(groups.index(group))
            print("found a group")
            
            break
        else:
            print("didn't find group")
```

```python
groupings
```

```python
df['groupID'] = groupings
```

```python
def groupFinder(item):
    for group in groups:
        if item in group:
            return groups.index(group)

df['groupID2'] = groupFinder(df['item'])
```
